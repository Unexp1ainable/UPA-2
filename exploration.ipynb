```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pprint import pprint
import scipy
```

```python
data = pd.read_csv("penguins_lter.csv")
plt.title("Histogram of species")
plt.hist(data["Species"])
plt.xticks(rotation=10)
```

```python
plt.title("Histogram over Island distribution")
plt.hist(data["Island"], color="purple")
plt.xticks(rotation=10)
```
